```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results.csv')
```

```python
df.drop(columns='Wall time').drop(columns = 'Step')
```

```python
ax = plt.gca()

df.plot(kind='line', y='train_acc',ax=ax)
df.plot(kind='line',y='val_acc', color='red', ax=ax)
ax.set_xlabel('Epoch')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy after 30 epochs')
plt.draw()
plt.savefig('Accuracy.png', dpi=100)
plt.show()
```

```python
ax = plt.gca()

df.plot(kind='line', y='train_loss',ax=ax)
df.plot(kind='line',y='val_loss', color='red', ax=ax)
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
ax.set_title('Loss after 30 epochs')
plt.draw()
plt.savefig('Loss.png', dpi=100)
plt.show()
```
